```python
sub_mean = df.groupby("gics_sub_industry").mean()
sub_mean.head()
```

```python
average_sub_industry = sub_mean["value"]
```

```python
average_sub_industry = average_sub_industry.to_frame()
```

```python
average_sub_industry.head()
```

```python
df2 = df.merge(average_sub_industry, on="gics_sub_industry")
```

```python
df2.rename(columns={"value_x": "ebita_multiple","value_y":"sub_industry_average"}, inplace=True)
```

```python
# df2["percent_change"] = ((df2["ebita_multiple"]-df2["sub_industry_average"])/df2["sub_industry_average"])*100
df2["percent_change"] = ((df2["sub_industry_average"]-df2["ebita_multiple"])/df2["ebita_multiple"])*100
    
    
```
