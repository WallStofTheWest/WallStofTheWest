```python
''' Function to clean the text by removing links and special characters using regex. '''
def clean_text(data):
    return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", data).split())
```

```python
def sa_analyzer(data):
    analysis = TextBlob(clean_text(data))
    if analysis.sentiment.polarity > 0:
        return 1
    if analysis.sentiment.polarity == 0:
        return 0
    else:
        return -1
```

```python
tweet_df['sentiment_analysis'] = np.array([sa_analyzer(item) for item in tweet_df['text']])
```
